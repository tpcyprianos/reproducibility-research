# Best Practices
The goal of this document is to present some importants tips about reproducible research. The description here shows what worked and what didn't work during my research to the course IA369-Z from FEEC Unicamp. Here you will find *my lessons learned* during the course, and I hope to help you with your research.

## Reproducible Research
To do reproducible research is something to do during your research, it is a way to do research, and it is not a step that you will do in the end of your research, like: "Ok, now I will put my files on Github and everybody will be able to reproduce my experiments". Reproducible research is a practical approach that will helps you to improve the **quality** of your work, to **validate** your finds and to **do relevant contributions in your area**. When others researchers can reproduce your experiment, they can use and create from your work. Even they find problems in your experiment, surely they will let you know, what is a good thing.

I thing, sometimes, we are so concentrate in the validation of our hypothesis, and in the writting (paper) of our insights that we forgot report detail steps and decisions (for us obvious), becoming the reproduction impossible (for other researchers and yourself).

What helped me during the experiments was keep in my mind *"A human will read my paper, code, and notes."*

If still are not convinced, please, read theses references: [Useless Studies](http://brasil.elpais.com/brasil/2017/01/10/internacional/1484073680_523691.html?id_externo_rsoc=FB_CC),  [Manifesto from Nature](https://www.nature.com/articles/s41562-016-0021)


## Resources for Reproducibility

Now, I hope, you are motivated to do reproducible research, then, you can read this [10 Simple Rules](http://journals.plos.org/ploscompbiol/article?id=10.1371/journal.pcbi.1003285). 

After discussions during the classes course, we agreed with following resources of Reproducible Research are:
* Version Control
* Workflow
* Data repository
* Code development
* Documentation
* Environment

Now, I will treat each element. For being clearer, I will separate the tips in DO (it works) and (DON'T) (it doesn't work):

### Workflow

**---- DO ----**
* For me, to design the workflow must be the first step of your research. Even in the beginning, it will help you to think and draw the flow of the research with the inputs, outputs, dependencies and data. Moreover, it will help to plan your research;
* Save your editable files (for any tool), certainly, the workflow will change during the process;
* [Draw.io](http://draw.io/) is a good tool and is integrated with Git (the integration works very well);
* 

**---- DON'T ----**
* Do not underestimate this step, put as much detail as possible. 

### Version Control

**---- DO ----**
* After plan you research, make a friendship with version control;
* [Git](https://git-scm.com/) is a good and disseminated option (and you will find a lot of documentation);
* Select a repository, [Github](https://github.com/) and [Bitbucket](https://bitbucket.org/) are famous;
* Use the version control DURING your work, making ``commits`` in changes that you want to save;
* If you new in git, there are some guides: [Try Git](https://try.github.io/), [Github Guides](https://guides.github.com/) 

**---- DON'T ----**
* Do not leave to the end of your research put your codes(and related files) in the repository. Make your repository in the beginning.
